This package consists of Keras layers for quantum classifier and quantum convolution, using "data re-uploading circuit" as the parameterized quantum circuit (PQC). For a more in-depth theoretical equation and design details of the data re-uploading circuit, it is highly recommended to read the [original paper](https://quantum-journal.org/papers/q-2020-02-06-226/). The paper is very well written (clear, concise, easy to understand).

In this tutorial, there will be no details about what is the physical or mathematical meaning for certain arguments, e.g., number of qubits, as it is already explained clearly in the original paper. This tutorial will only explain about how to use the package. It is assumed that the readers are already familiar with the data re-uploading circuit.

In this first section, the quantum classifier layer and quantum convolution layer will be explained.

## 1. Quantum Classifier with Data Re-uploading PQC

`ReUploadingPQC` class is a Keras layer for the quantum classifier. It is very similar to the fully-connected layer, or known as [Dense layer](https://keras.io/api/layers/core_layers/dense/) in Keras, for classical neural networks.

It receives 2D `(num. samples, num. features)` input vector and will return 2D `(num. samples, N)` output vector. `N` is the number of expectation values of observables measured at the end of the circuit.

### Arguments:
- **n_qubits**: An `integer`, number of qubits to be used for the data re-uploading PQC.
- **n_layers**: An `integer`, number of layers to be used for the data re-uploading PQC.
- **input_size**: An `integer`, number of features of the input vector `(num. features)`.
- **use_entanglement**: A `boolean` to determine whether to use the entangling layer between the rotational layers. The entangling layer is a layer of CZ entangling gates on all qubits, arranged in a circular topology. `Default: True`.
- **use_terminal_entanglement**: A `boolean` to determine whether to use the entangling layer after the last rotational layer. `Default: True`.
- **observables**: A `list` of [Cirq operators](https://quantumai.google/cirq/operators_and_observables) that define measurements to be done on the circuit. If `None`, the last qubit of the circuit will be measured with a Pauli-Z gate as the observable. `Default: None`.
- **name**: A string, used as the name of the layer. When constructing a Keras model with more than one of this layer, the `name` must be specified differently between layers. `Default: "re-uploading_PQC"`.

### Attributes:
- **n_qubits**: (see **Arguments**).
- **n_layers**: (see **Arguments**).
- **input_size**: (see **Arguments**).
- **use_entanglement**: (see **Arguments**).
- **use_terminal_entanglement**: (see **Arguments**).
- **main_name**: same as `name` in **Arguments**.
- **qubits**: [`Cirq GridQubit`](https://quantumai.google/cirq/qubits), `cirq.GridQubit.rect(1, n_qubits)`.
- **observables**: (see **Arguments**).
- **circuit**: [`Cirq Circuit`](https://quantumai.google/cirq/circuits).
- **thetas**: `TensorFlow Variable`, the $\theta$ trainable variables inside the data re-uploading circuit, act similar (but not the same) to bias in a classical fully-connected layer.
- **w**: `TensorFlow Variable`, the **weights** trainable variables inside the data re-uploading circuit, act similar to weights in a classical fully-connected layer.
- **indices**: `TensorFlow Constant`, a set of indices that maps the `SymPy symbols` to the corresponding `TensorFlow Variable`.
- **empty_circuit**: `TensorFlow Constant`, TensorFlow conversion of an empty `Cirq Circuit` to act as a "hack" to implement the data re-uploading circuit as a [TFQ ControlledPQC class](https://www.tensorflow.org/quantum/api_docs/python/tfq/layers/ControlledPQC).
- **computation_layer**: `TFQ ControlledPQC class`.

### Methods:
- **get_config**: `get_config()`, the [TensorFlow Keras Layers `get_config` method](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Layer#get_config). This is needed if we want to make a custom layer that is saveable via [`tf.train.Checkpoint`](https://www.tensorflow.org/api_docs/python/tf/train/Checkpoint).
- **call**: `call(inputs)`, the [TensorFlow Keras Layers `call` method](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Layer#call).

### Some Examples:

Below is some usage examples.<br>
Note that a more complete examples on how to train a quantum classifier model can be found in the [Tutorial](https://colab.research.google.com/drive/1TWQgdVzKUITMNzIUFtyzq9MfuTfkyJVe?usp=sharing).

Example 1:

A 2-qubits and 1-layer data re-uploading PQC without any entanglement, measured by a $|0⟩⟨0|$ on the first qubit and $|1⟩⟨1|$ on the second qubit. The output of this layer is a vector with dimension `(num. samples, 2)`.

Let's say the input vector to this layer has shape of `(num. samples, 9)`.

````python
n_qubits = 2
n_layers = 1
input_size = 9
qubits = cirq.GridQubit.rect(1, n_qubits)

Z_0 = cirq.PauliString(cirq.Z(qubits[0]))
I_0 = cirq.PauliString(cirq.I(qubits[0]))
Z_1 = cirq.PauliString(cirq.Z(qubits[1]))
I_1 = cirq.PauliString(cirq.I(qubits[1]))

gate_0 = 0.5*Z_0 + 0.5*I_0  # equivalent to |0⟩⟨0|
gate_1 = -0.5*Z_1 + 0.5*I_1 # equivalent to |1⟩⟨1|

observables = [gate_0, gate_1]

PQC = ReUploadingPQC(n_qubits, n_layers, input_size, use_entanglement=False, use_terminal_entanglement=False, observables=observables)
SVGCircuit(PQC.circuit)
````

````python
# Check the size of the output

input_tensor = tf.keras.Input(shape=(input_size, 1), dtype=tf.dtypes.float32, name='input')
PQC(input_tensor).shape
````

Example 2:

A 4-qubits and 2-layers data re-uploading PQC with entanglement but without the last entangling layer, measured by a Pauli-Z operator on each qubit. The output of this layer is a vector with dimension `(num. samples, 4)`.

Let's say the input vector to this layer has shape of `(num. samples, 5)`.

````python
n_qubits = 4
n_layers = 2
input_size = 5
qubits = cirq.GridQubit.rect(1, n_qubits)

observables = [cirq.Z(qubits[i]) for i in range(n_qubits)]

PQC = ReUploadingPQC(n_qubits, n_layers, input_size, use_entanglement=True, use_terminal_entanglement=False, observables=observables)
SVGCircuit(PQC.circuit)
````

````python
# Check the size of the output

input_tensor = tf.keras.Input(shape=(input_size, 1), dtype=tf.dtypes.float32, name='input')
PQC(input_tensor).shape
````

Example 3:

A 3-qubits and 3-layers data re-uploading PQC with full entanglement (including the last layer), measured by a $Z⊗Z⊗Z$ operator. The output of this layer is a vector with dimension `(num. samples, 1)`.

Let's say the input vector to this layer has shape of `(num. samples, 4)`.

````python
n_qubits = 3
n_layers = 3
input_size = 4
qubits = cirq.GridQubit.rect(1, n_qubits)

gates = [cirq.PauliString(cirq.Z(qubits[i])) for i in range(n_qubits)]

observables = gates[0]*gates[1]*gates[2]

PQC = ReUploadingPQC(n_qubits, n_layers, input_size, use_entanglement=True, use_terminal_entanglement=True, observables=observables)
SVGCircuit(PQC.circuit)
````

````python
# Check the size of the output

input_tensor = tf.keras.Input(shape=(input_size, 1), dtype=tf.dtypes.float32, name='input')
PQC(input_tensor).shape
````

## 2. Quantum Convolution with Data Re-uploading PQC

The difference between the quantum convolution layer with the classical one is that we used a trainable variational quantum circuit instead of a classical filter to produce the feature map.

This class utilizes the `ReUploadingPQC` class to build the convolutional layer. In this package's implementation, we use the data re-uploading circuit as the trainable variational quantum circuit for the filter.

For example, if you define a quantum convolution layer with 4 output filters, and it receives RGB image as input (3 input channels), then there will be $3 \times 4 = 12$ data re-uploading PQCs that act as the filters.

All of the data re-uploading PQCs will have the default observable for measurements, which is a Pauli-Z gate on the last qubit.

example.png

`QConv2D_DRC` class is a class similar to a Keras layer (not the same) for the quantum classifier. It is very similar to the convolutional layer, or known as [Conv2D layer](https://keras.io/api/layers/convolution_layers/convolution2d/) in Keras, for classical CNN.

It receives 4D `(num. samples, input rows, input cols, num. channels)` input vector and will return 4D `(num. samples, output rows, output cols, num. output filters)` output vector.

### Arguments:
- **filters**: An `integer`, the dimensionality of the output space (i.e. the number of output filters in the convolution).
- **kernel_size**: An integer or tuple/list of 2 integers, specifying the height and width of the 2D convolution window. Can be a single integer to specify the same value for all spatial dimensions.
- **strides**: An integer or tuple/list of 2 integers, specifying the strides of the convolution along the height and width. Can be a single integer to specify the same value for all spatial dimensions.
- **drc_hyperparameters**: A `dictionary`, listing important arguments for the `ReUploadingPQC` class. See more details in the cell below.
- **layer_id**: An `integer`, used for the name of the layer. When constructing a Keras model with more than one of this layer, the `layer_id` must be specified differently between layers.
- **padding**: one of `None` or `"valid"` (case-sensitive). `None` means no padding, some of the input pixels might not be used (discarded) because of the convolution operation. `"valid"` means padding the input image with zeros so that all of the input pixels are guaranteed to be used in the convolution operation.

Notes on **drc_hyperparameters**:

```
drc_hyperparameters = {
    "n_qubits": n_qubits_list,
    "n_layers": n_layers_list,
    "use_ent": use_ent_list,
    "use_terminal_ent": use_terminal_ent_list
}
```
- **n_qubits_list**: `list` of integers to specify the number of qubits to be used for the data re-uploading PQC.
- **n_layers_list**: `list` of integers to specify the number of layers to be used for the data re-uploading PQC.
- **use_ent_list**: `list` of booleans to determine whether to use the entangling layer between the rotational layers. The entangling layer is a layer of CZ entangling gates on all qubits, arranged in a circular topology.
- **use_terminal_ent_list**: `list` of booleans to determine whether to use the entangling layer after the last rotational layer.

The `len` of every list must equal to the number of input channels, e.g., the `n_qubits_list[0]` will be applied to the data re-uploading PQC that acts as a filter for the first input channel.

### Attributes:
- **filters**: (see **Arguments**).
- **layer_id**: (see **Arguments**).
- **padding**: (see **Arguments**).
- **n_qubits**: `drc_hyperparameters["n_qubits"]`, (see **Arguments**).
- **n_layers**: `drc_hyperparameters["n_layers"]`, (see **Arguments**).
- **use_ent**: `drc_hyperparameters["use_ent"]`, (see **Arguments**).
- **use_terminal_ent** = `drc_hyperparameters["use_terminal_ent"]`, (see **Arguments**).
- **h_kernel_size**: `kernel_size[0]` or `kernel_size`, (see **Arguments**).
- **w_kernel_size**: `kernel_size[1]` or `kernel_size`, (see **Arguments**).
- **h_stride**: `strides[0]` or `strides`, (see **Arguments**).
- **w_stride**: `strides[1]` or `strides`, (see **Arguments**).

### Methods:
- **call**: `call(inputs)`
  - Args:
    - `inputs`: Input tensor with shape `num. samples + (input rows, input cols, num. channels)`. The first positional inputs argument is subject to special rules:
       - inputs must be explicitly passed. A layer cannot have zero arguments, and inputs cannot be provided via the default value of a keyword argument.
       - NumPy array or Python scalar values in inputs get cast as tensors.
       - Layers are built (build(input_shape) method) using shape info from inputs only.
  - Returns: A tensor of rank 4 with shape `num. samples + (output rows, output cols, num. output filters)`.

### Some Examples:

Below is some usage examples.<br>
Note that a more complete examples on how to train a model with a quantum convolution layer can be found in the [Tutorial](https://colab.research.google.com/drive/1TWQgdVzKUITMNzIUFtyzq9MfuTfkyJVe?usp=sharing).

Example 1:

A quantum convolution layer with 2 output filters, kernel size of 3, strides of 2, and no padding. The `drc_hyperparameters` are as defined in the code block below.

Let's say the input image to this layer has shape of `(num. samples, 28, 28, 3)`.

````python
input_image_shape = (28,28,3)

filters = 2
kernel_size = 3
strides = 2

input_tensor = tf.keras.Input(shape=input_image_shape, dtype=tf.dtypes.float32, name='input')

drc_hyperparameters = {
    "n_qubits": [1, 1, 2],
    "n_layers": [1, 1, 1],
    "use_ent": [True, True, True],
    "use_terminal_ent": [True, True, True]
}

qconv = QConv2D_DRC(filters, kernel_size, strides, drc_hyperparameters, layer_id=1, padding=None).call(input_tensor)
````

````python
# Check the size of the output

qconv.shape
````

Example 2:

A quantum convolution layer with 5 output filters, kernel size of 3 (row axis) and 2 (column axis), strides of 3 (row axis) and 2 (column axis), and "valid" padding. The `drc_hyperparameters` are as defined in the code block below.

Let's say the input image to this layer has shape of `(num. samples, 15, 15, 1)`.

````python
input_image_shape = (15,15,1)

filters = 5
kernel_size = (3, 2)
strides = (3, 2)

input_tensor = tf.keras.Input(shape=input_image_shape, dtype=tf.dtypes.float32, name='input')

drc_hyperparameters = {
    "n_qubits": [1, 1, 2, 3, 3],
    "n_layers": [1, 2, 1, 2, 1],
    "use_ent": [True, True, True, False, False],
    "use_terminal_ent": [True, True, True, False, False]
}

qconv = QConv2D_DRC(filters, kernel_size, strides, drc_hyperparameters, layer_id=1, padding="valid").call(input_tensor)
````

````python
# Check the size of the output

qconv.shape
````
